```python
async def get_stadium_data(stadium_data) -> dict:
    data = {}
    td_list = stadium_data.findAll('td')
    data['country'] = 'Brasil'
    data['stadium_name'] = td_list[0].find('a').text
    data['city'] = td_list[0].find('span').text
    data['state'] = ''
    if '(' in data['city']:
        data['state'] = data['city'].split('(')[1].split(')')[0]
        data['city'] = data['city'].split('(')[0].strip()
    capacity = str(td_list[1]).split('Lotação:')[1].split('<br/>')[0].strip()
    data['capacity'] = '' if capacity == '-' else int(capacity)
    creation_year = td_list[1].find('span').text.split(':')[1].strip()
    data['creation_year'] = None if creation_year == '-' else int(creation_year)
    data['team'] = td_list[2].find('a').text if '</a>' in str(td_list[2]) else ''

    return data
    
```

```python
import asyncio
tr_list = soup.find('div', id='page_main', class_='home').find_all("tr")[1:-1]
    
# Create a list of coroutines
coroutines = [get_stadium_data(tr) for tr in tr_list]

# Execute the coroutines concurrently
list_data = []
list_data += await asyncio.gather(*coroutines)
```
